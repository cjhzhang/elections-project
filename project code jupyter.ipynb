```r
elec_data <- read.csv('anes_timeseries_cdf_csv_20220916.csv')
#having trouble uploading it to github rn

subset <- elec_data[, c("VCF0004", "VCF0114", "VCF0110", "VCF0104", "VCF0105a", "VCF0102","VCF0705","VCF0901a")]
subset <- na.omit(subset)
colnames(subset)[1]<-'year_of_study' #was VCF0004
colnames(subset)[2]<-'income_group' #was VCF0114
colnames(subset)[3]<-'education_4_cat' #was VCF0110
colnames(subset)[4]<-'gender' #was VCF0104
colnames(subset)[5]<-'race_ethnicity_seven' #was VCF0105a
colnames(subset)[6]<-'age_group' #was VCF0102
colnames(subset)[7]<-'vote_pres' #was VCF0705
colnames(subset)[8] <- "state_code" #was VCF0901a, state codes can be found on page 101 of appendices

subset
```

```r
#converting ints into factors for categorical variables, state code into character for converting to state name
subset$income_group <- as.factor(subset$income_group)
subset$education_4_cat <- as.factor(subset$education_4_cat)
subset$gender <- as.factor(subset$gender)
subset$race_ethnicity_seven <- as.factor(subset$race_ethnicity_seven)
subset$age_group <- as.factor(subset$age_group)
subset$vote_pres <- as.factor(subset$vote_pres)
subset$state_code <- as.character(subset$state_code)
```

```r
#filtering for after 1980, not including 1980
post_1980 <- filter(subset, year_of_study > 1980)

#if we want to filter out all missing, i have included the code to rename
#the missing levels as comments
post_180_nomissing <- filter(post_1980, 
                       income_group != 0, 
                       education_4_cat != 0, 
                       gender != 0,
                       race_ethnicity_seven != 9,
                       age_group != 0,
                       vote_pres != 0,
                            vote_pres != 3, #remove 3rd party 
                            state_code != 99,
                            state_code != 00)

```
