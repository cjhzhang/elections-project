```r
library(readr)
library(dplyr)
library(lubridate)
library(tidyverse)
library(vtable)
library(stargazer)
```

```r
elec_data <- read.csv('anes_timeseries_cdf_csv_20220916.csv')
```

```r
subset <- elec_data[, c("VCF0004", "VCF0114", "VCF0104","VCF0705", 'VCF0105a','VCF0102')]
subset <- na.omit(subset)
colnames(subset)[1]<-'year_of_study' #was VCF0004
colnames(subset)[2]<-'income_group' #was VCF0114
colnames(subset)[3]<-'gender' #was VCF0104
colnames(subset)[4]<-'vote_pres' #was VCF0705
colnames(subset)[5]<-'race_ethnicity_seven' #was VCF0105a
colnames(subset)[6]<-'age_group' #was VCF0102

subset
```

```r
#filtering for after 1980, not including 1980
post_1980 <- filter(subset, year_of_study > 1980)


post_180_nomissing <- filter(post_1980, 
                       income_group != 0, 
                       gender != 0,
                       vote_pres != 0,
                        vote_pres != 3,
                        race_ethnicity_seven != 9,
                       age_group != 0)
```

```r
#renaming
#income groups
post_180_nomissing$income_group[post_180_nomissing$income_group==1] <- "0-16 percentile"
post_180_nomissing$income_group[post_180_nomissing$income_group==2] <- "17-33 percentile"
post_180_nomissing$income_group[post_180_nomissing$income_group==3] <- "34-67 percentile"
post_180_nomissing$income_group[post_180_nomissing$income_group==4] <- "68-95 percentile"
post_180_nomissing$income_group[post_180_nomissing$income_group==5] <- "95-100 percentile"
#gender
post_180_nomissing$gender[post_180_nomissing$gender==1] <- "Male"
post_180_nomissing$gender[post_180_nomissing$gender==2] <- "Female"
post_180_nomissing$gender[post_180_nomissing$gender==3] <- "Other"
#race
post_180_nomissing$race_ethnicity_seven[post_180_nomissing$race_ethnicity_seven==1] <- "White"
post_180_nomissing$race_ethnicity_seven[post_180_nomissing$race_ethnicity_seven==2] <- "Black"
post_180_nomissing$race_ethnicity_seven[post_180_nomissing$race_ethnicity_seven==3] <- "Asian or Pacific Islander"
post_180_nomissing$race_ethnicity_seven[post_180_nomissing$race_ethnicity_seven==4] <- "American Indian or Alaska Native"
post_180_nomissing$race_ethnicity_seven[post_180_nomissing$race_ethnicity_seven==5] <- "Hispanic"
post_180_nomissing$race_ethnicity_seven[post_180_nomissing$race_ethnicity_seven==6] <- "Other or Multiple"
#age
post_180_nomissing$age_group[post_180_nomissing$age_group==1] <- "17-24"
post_180_nomissing$age_group[post_180_nomissing$age_group==2] <- "25-34"
post_180_nomissing$age_group[post_180_nomissing$age_group==3] <- "35-44"
post_180_nomissing$age_group[post_180_nomissing$age_group==4] <- "45-54"
post_180_nomissing$age_group[post_180_nomissing$age_group==5] <- "55-64"
post_180_nomissing$age_group[post_180_nomissing$age_group==6] <- "65-74"
post_180_nomissing$age_group[post_180_nomissing$age_group==7] <- "75-99+"
#votepres
post_180_nomissing$vote_pres[post_180_nomissing$vote_pres==1] <- "democrat"
post_180_nomissing$vote_pres[post_180_nomissing$vote_pres==2] <- "republican"

#converting to factors
post_180_nomissing$vote_pres <- as.factor(post_180_nomissing$vote_pres)
post_180_nomissing$income_group <- as.factor(post_180_nomissing$income_group)
post_180_nomissing$gender <- as.factor(post_180_nomissing$gender)
post_180_nomissing$age_group <- as.factor(post_180_nomissing$age_group)
post_180_nomissing$race_ethnicity_seven <- as.factor(post_180_nomissing$race_ethnicity_seven)
```

```r
#table_sum_stats <- sumtable(post_180_nomissing, out='latex')
#uncomment when we want the latex output, it's super long 
table_sum_stats_df <- sumtable(post_180_nomissing, out='return')
table_sum_stats_df
```

```r
post_180_nomissing$vote_pres <- as.character(post_180_nomissing$vote_pres)

dem_df <- post_180_nomissing
dem_df$vote_pres[dem_df$vote_pres=='republican'] <- 0
dem_df$vote_pres[dem_df$vote_pres=='democrat'] <- 1
dem_df$vote_pres <- as.integer(dem_df$vote_pres)

rep_df <- post_180_nomissing
rep_df$vote_pres[rep_df$vote_pres=='democrat'] <- 0
rep_df$vote_pres[rep_df$vote_pres=='republican'] <- 1

dem_df
```

```r
allvote_dem_og <- lm(vote_pres ~ income_group + gender, data = dem_df)
allvote_dem_age <- lm(vote_pres ~ income_group + gender + age_group, data = dem_df)
allvote_dem_race <- lm(vote_pres ~ income_group + gender + race_ethnicity_seven, data = dem_df)

allvote_rep_og <- lm(vote_pres ~ income_group + gender, data = rep_df)
allvote_rep_age <- lm(vote_pres ~ income_group + gender + age_group, data = rep_df)
allvote_rep_race <- lm(vote_pres ~ income_group + gender + race_ethnicity_seven, data = rep_df)
```
