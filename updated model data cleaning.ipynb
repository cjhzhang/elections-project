```r
library(readr)
library(dplyr)
library(lubridate)
library(tidyverse)
library(vtable)
```

```r
elec_data <- read.csv('anes_timeseries_cdf_csv_20220916.csv')
```

```r
subset <- elec_data[, c("VCF0004", "VCF0114", "VCF0104","VCF0705")]
subset <- na.omit(subset)
colnames(subset)[1]<-'year_of_study' #was VCF0004
colnames(subset)[2]<-'income_group' #was VCF0114
colnames(subset)[3]<-'gender' #was VCF0104
colnames(subset)[4]<-'vote_pres' #was VCF0705


subset
```

```r
subset$income_group <- as.factor(subset$income_group)
subset$gender <- as.factor(subset$gender)
subset$vote_pres <- as.factor(subset$vote_pres)
str(subset)
```

```r
#filtering for after 1980, not including 1980
post_1980 <- filter(subset, year_of_study > 1980)


post_180_nomissing <- filter(post_1980, 
                       income_group != 0, 
                       gender != 0,
                       vote_pres != 0,
                        vote_pres != 3)

str(post_180_nomissing)
```

```r
gendermodel <- lm(vote_pres ~ income_group + gender, data = post_180_nomissing)
summary(gendermodel)
```

```r
table_sum_stats <- sumtable(post_180_nomissing, out='latex')

table_sum_stats_df <- sumtable(post_180_nomissing, out='return')
table_sum_stats_df
```
